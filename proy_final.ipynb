```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
from sklearn.metrics import confusion_matrix, accuracy_score
import sklearn.metrics as metrics
```

```python
df = pd.read_csv('Weather_Data.csv')
df.head()
```

```python
df_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
df_processed.replace(['No', 'Yes'], [0,1], inplace=True)
df_processed.drop('Date', axis=1, inplace=True)
df_processed = df_processed.astype(float)
df_processed.head()
```

```python
x = df_processed.drop(columns='RainTomorrow', axis=1)
y = df_processed['RainTomorrow'] 
```

Linear Regression

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
LinearReg = LinearRegression().fit(x_train, y_train)
predictions = LinearReg.predict(x_test)
```

```python
from sklearn.metrics import r2_score

LinearRegression_MAE = np.mean(np.absolute(predictions - y_test))
LinearRegression_MSE = np.mean((predictions - y_test) **2)
LinearRegression_R2 = r2_score(y_test, predictions)
```

```python
report = pd.DataFrame({
    'MAE': [LinearRegression_MAE],
    'MSE': [LinearRegression_MSE],
    'R2': [LinearRegression_R2]
})

report.head()
```

```python
KNN = KNeighborsClassifier(n_neighbors=4).fit(x_train, y_train)
predictions = KNN.predict(x_test)
```

```python
KNN_Accuracy_Score = accuracy_score(y_test, predictions)
KNN_JaccardIndex = jaccard_score(y_test, predictions)
KNN_F1_Score = f1_score(y_test,predictions)

report = pd.DataFrame({
    'Accuracy Score': [KNN_Accuracy_Score],
    'Jaccar Score': [KNN_JaccardIndex],
    'F1 Score': [KNN_F1_Score]
})

report.head()
```

Decision Tree

```python
Tree = DecisionTreeClassifier(criterion='entropy', max_depth=4).fit(x_test, y_test)
predictions = Tree.predict(x_test)
```

```python
Tree_Accuracy_Score = accuracy_score(y_test, predictions)
Tree_JaccardIndex =  jaccard_score(y_test, predictions)
Tree_F1_Score = f1_score(y_test, predictions)

report = pd.DataFrame({
    'Accuracy Score': [Tree_Accuracy_Score],
    'Jaccar Score': [Tree_JaccardIndex],
    'F1 Score': [Tree_F1_Score]
})

report.head()
```

Regresion logistica


```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
LR = LogisticRegression(solver='liblinear').fit(x_train, y_train)
```

```python
predictions = LR.predict(x_test)
predictions_proba = LR.predict_proba(x_test)
```

```python
LR_Accuracy_Score = accuracy_score(y_test, predictions)
LR_JaccardIndex = jaccard_score(y_test, predictions)
LR_F1_Score = f1_score(y_test, predictions)
LR_Log_Loss = log_loss(y_test, predictions_proba)

report = pd.DataFrame({
    'Accuracy Score': [LR_Accuracy_Score],
    'Jaccar Score': [LR_JaccardIndex],
    'F1 Score': [LR_F1_Score],
    'Loss Log' : [LR_Log_Loss]
})

report.head()
```

Support Vector Machine

```python
SVM = svm.SVC().fit(x_train, y_train)
```

```python
predictions = SVM.predict(x_test)
```

```python
SVM_Accuracy_Score = accuracy_score(y_test, predictions)
SVM_JaccardIndex = jaccard_score(y_test, predictions)
SVM_F1_Score = f1_score(y_test, predictions)

report = pd.DataFrame({
    'Accuracy Score': [SVM_Accuracy_Score],
    'Jaccar Score': [SVM_JaccardIndex],
    'F1 Score': [SVM_F1_Score],
})

report.head()
```

```python
report = pd.DataFrame({
    'Model': ['SVM', 'KNN', 'Decision Tree', 'Logistic Regression'],
    'Accuracy Score': [SVM_Accuracy_Score, KNN_Accuracy_Score, Tree_Accuracy_Score, LR_Accuracy_Score],
    'Jaccar Score': [SVM_JaccardIndex, KNN_JaccardIndex, Tree_JaccardIndex, LR_JaccardIndex],
    'F1 Score': [SVM_F1_Score, KNN_F1_Score, Tree_F1_Score, LR_F1_Score],
    'LOG LOSS' : [None,None,None,LR_Log_Loss]
})

report.head()
```
